```python
import cobra
```

```python
from cobra import Model,Reaction,Metabolite
```

```python
model=Model("ex")
```

```python
v1=Reaction("v1")
v1.name="v1"
v1.lower_bound=0
v1.upper_bound=1000
```

```python
v2=Reaction("v2")
v2.name="v2"
v2.lower_bound=0
v2.upper_bound=1000
```

```python
v0=Reaction("v0")
v0.name="v0"
v0.lower_bound=1
v0.upper_bound=1
```

```python
v3=Reaction("v3")
v3.name="v3"
v3.lower_bound=0
v3.upper_bound=1000
```

```python
ATP_R=Reaction("ATP_R")
ATP_R.name="ATP"
ATP_R.lower_bound=0.9
ATP_R.upper_bound=0.9
```

```python
M=Reaction("M")
M.name="M"
M.lower_bound=0
M.upper_bound=1000
```

```python
A=Metabolite('A',compartment="c")
B=Metabolite('B',compartment="c")
C=Metabolite('C',compartment="c")
ATP=Metabolite('ATP',compartment="c")
```

```python
v1.add_metabolites({A:-1,B:1})
```

```python
v2.add_metabolites({B:-1,C:1})
```

```python
v0.add_metabolites({A:1})
```

```python
v3.add_metabolites({A:-1,ATP:1})
ATP_R.add_metabolites({ATP:1})
M.add_metabolites({C:-1})
```

```python
model.add_reactions([v0,v1,v2,v3,M])
```

```python
model.objective='M'
```

```python
model.optimize()
```
